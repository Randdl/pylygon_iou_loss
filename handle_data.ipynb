```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
variance = 20
setting = "lr2e-1v{}".format(variance)
total_iter_poly_hist = np.load("results/poly_hist_{}.npy".format(setting))
total_iter_l1_hist = np.load("results/l1_hist_{}.npy".format(setting))
total_iter_l2_hist = np.load("results/l2_hist_{}.npy".format(setting))
plt.close()
kwargs = dict(alpha=0.5, bins=20, stacked=True)
plt.title('Variance = {} Learning rate = 0.2'.format(variance))
plt.hist(total_iter_poly_hist[total_iter_poly_hist < 600], **kwargs, color='g', label='poly')
plt.hist(total_iter_l1_hist, **kwargs, color='b', label='l1')
plt.hist(total_iter_l2_hist, **kwargs, color='r', label='l2')
plt.legend()
plt.show()
```
